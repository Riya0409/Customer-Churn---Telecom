```python
# TotalCharges  --> is in string --> convert to numeric

customer_churn["TotalCharges"] = customer_churn["TotalCharges"].replace(" ", np.nan)
```

```python
customer_churn["TotalCharges"] = customer_churn["TotalCharges"].fillna(customer_churn["TotalCharges"].mode()[0])
```

```python
customer_churn["TotalCharges"] = customer_churn["TotalCharges"].astype(float)
```

```python
customer_churn = customer_churn.drop("customerID", axis=1)
```

```python
customer_churn.columns
```

```python
customer_churn = pd.get_dummies(customer_churn, columns= ["gender","Partner","Dependents",'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod' ], drop_first=True)
```

```python
customer_churn.info()
```

```python
customer_churn.head()
```

```python
## converting boolean into integer

df = customer_churn.applymap(lambda x : int(x) if isinstance(x,bool) else x)
```

```python
df.head()
```

```python
## Split the data into x and y

x = df.drop(["Churn"], axis= 1)
y = df["Churn"]
```

```python
x
```

```python
y
```

```python
## feature Scaling Required

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler_columns = ['tenure','MonthlyCharges','TotalCharges']

x[scaler_columns] = scaler.fit_transform(x[scaler_columns])
```

```python
x.head()
```

```python
## Handling imbalance daatset by using SMOTE techniques

import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE()

x_smote,y_smote = smote.fit_resample(x,y)
print(y.value_counts())
print()
print(y_smote.value_counts())
```

```python
## Spliting data into training and testing

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_smote,y_smote, test_size=0.3, random_state= 101)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(x_train,y_train)
```

```python
y_pred_train = logit.predict(x_train)
y_pred_test = logit.predict(x_test)
```

```python
## Evaluation Metrics

from sklearn.metrics import classification_report, accuracy_score
```

```python
print(classification_report(y_train,y_pred_train))
print()
print(classification_report(y_test,y_pred_test))
```

```python

print(accuracy_score(y_train,y_pred_train))
print()
print(accuracy_score(y_test,y_pred_test))
```
